To check how old the car is, lets create a feature car age

```python
df['Current_Year'] = 2020
df['No_of_years'] = df['Current_Year'] - df['Year']
df.head()
```

```python
df_sorted = df.sort_values(by='No_of_years', ascending=False)
top_5_cars = df_sorted.head(5)
print(top_5_cars[['Car_Name', 'Year', 'No_of_years']])
```

```python
import matplotlib.pyplot as plt
```

```python
colors = ['blue', 'green', 'orange', 'red', 'purple']

plt.figure(figsize=(10, 6))
plt.bar(top_5_cars['Car_Name'], top_5_cars['No_of_years'], color=colors)
plt.xlabel('Car Name')
plt.ylabel('Number of Years')
plt.title('Top 5 Cars with the Most Number of Years')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```

```python
df.drop(['Year'], axis=1, inplace=True)
df.drop(['Current_Year'], axis=1, inplace=True)
df.drop(['Car_Name'], axis=1, inplace=True)
df.head()
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X, y)
```

Importance of all the columns

```python
print(model.feature_importances_)  
print("highest", max(model.feature_importances_))
print("lowest", min(model.feature_importances_))
```

```python
pd.DataFrame(X.columns, model.feature_importances_)
```

Bar Graph

```python
importance = pd.Series(model.feature_importances_, index=X.columns)
top_features = importance.nlargest(8)
colors = ['blue', 'green', 'orange', 'red', 'purple', 'cyan', 'magenta', 'yellow']

plt.figure(figsize=(9, 5))
top_features.plot(kind='barh', color=colors)
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Top 8 Important Features')
plt.tight_layout()

plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
import numpy as np
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
print(n_estimators)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5,30, num=6)]
min_samples_split = [2,5,10,15,100]
min_samples_leaf = [1,2,5,10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
predictions = rf_random.predict(X_test)
predictions
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```

Accuracy

```python
from sklearn.metrics import r2_score
r2_score(y_test, predictions)
```
